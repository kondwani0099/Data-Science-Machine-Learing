**1. Print Statement**

Syntax: print("text or variables")
Explanation: Displays output to the console. You can use plain text, variables, or formatted strings to combine values.
Example:

```python
# Print a simple message
print("Hello, Welcome to Lesson 1!")
```

**2. Variables and Data Types**

Syntax: variable_name = value
Explanation: Variables store data values, and Python determines the type automatically (e.g., int, float, str).

```python
# Variables and their types
name = "Alice"         # String
age = 25               # Integer
height = 5.4           # Float
is_student = True      # Boolean

# Print variable values
print(f"Name: {name}, Age: {age}, Height: {height}, Student: {is_student}")
```

**3. Basic Input and Output**

Syntax: input("prompt")
Explanation: Takes input from the user as a string. Use print() to display a message or result.
Example:

```python
# Input from user
name = input("Enter your name: ")
print(f"Hello, {name}!")
```

**4. Conditional Statements**

Explanation: Executes code blocks based on conditions. Python uses indentation (not brackets) to define blocks.
Example:

```python
# If-else example
age = int(input("Enter your age: "))

if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")
```

**5.Loops**

Iterates over a sequence (e.g., range, list).

```python
# For loop example
for i in range(5):
    print(f"Loop iteration: {i}")

# While loop example
count = 0
while count < 3:
    print(f"Count is: {count}")
    count += 1
```

**6. Functions**

Explanation: Defines reusable code blocks. Use def to define a function and return to send back a result.
Example:

```python
# Function definition and call
def greet(name):
    return f"Hello, {name}!"

print(greet("Alice"))
```

**7. Lists**

Syntax: list_name = [item1, item2, ...]
Explanation: Stores an ordered collection of items. Lists are mutable (changeable).
Example:

```python
# List operations
fruits = ["apple", "banana", "cherry"]
fruits.append("orange")
fruits.remove("banana")

for fruit in fruits:
    print(fruit)
```

**8. Dictionaries**

Explanation: Stores key-value pairs. Access values using their keys.
Example:

```python
# Dictionary example
person = {
    "name": "Alice",
    "age": 25,
    "city": "New York"
}

# Accessing dictionary elements
print(person["name"])
print(person.get("age"))
```

**9. Classes and Objects**

Explanation: Python supports Object-Oriented Programming (OOP). Classes define blueprints for objects, and __init__ initializes object attributes.
Example:

```python
# Class definition and instantiation
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        return f"Hi, I'm {self.name} and I'm {self.age} years old."

alice = Person("Alice", 25)
print(alice.greet())
```

**10. File Handling**

Explanation: Handles files (e.g., reading or writing). Use with to ensure files close properly. Modes include r (read), w (write), and a (append).
Example:

```python
# Writing to a file
with open("example.txt", "w") as file:
    file.write("Hello, Colab!")

# Reading from a file
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

**11. Exception Handling**

Explanation: Manages runtime errors gracefully to prevent program crashes.
Example:

```python
# Try-except example
try:
    num = int(input("Enter a number: "))
    print(f"Number is: {num}")
except ValueError:
    print("That's not a valid number!")
```

**12. Importing Modules**

Syntax: import module_name
Explanation: Adds external functionality by importing libraries.
Example:

```python
# Using a module
import math

num = 16
print(f"Square root of {num} is {math.sqrt(num)}")
```

**13. List Comprehension**

Syntax: [expression for item in iterable]
Explanation: Creates lists concisely with a single line of code.
Example:

```python
# List comprehension example
squares = [x**2 for x in range(10)]
print(squares)
```

**14. Lambda Functions**

Syntax: lambda arguments: expression
Explanation: Creates small, anonymous functions for short-term use.
Example:

```python
# Lambda function example
square = lambda x: x**2
print(square(5))
```

**15. Working with NumPy (for Data Science)**

Explanation: NumPy is a library for numerical operations, offering fast array processing.
Example:

```python
import numpy as np

# Creating an array
arr = np.array([1, 2, 3, 4, 5])
print(f"Array: {arr}")
print(f"Mean: {np.mean(arr)}, Sum: {np.sum(arr)}")
```

```python
import numpy as np
arr = np.array([1, 2, 3])
print(arr.sum())
```

**16. Plotting with Matplotlib**

Explanation: Used for creating visualizations like line plots, bar charts, etc.
Example:

```python
import matplotlib.pyplot as plt

# Simple line plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y, marker='o')
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

**17. Pandas for Data Manipulation**

Explanation: Handles structured data using DataFrames for easy manipulation and analysis.
Example:

```python
import pandas as pd

# Creating a DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "Los Angeles", "Chicago"]
}

df = pd.DataFrame(data)
print(df)

# Filtering data
print(df[df["Age"] > 25])
```
